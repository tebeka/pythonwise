# matploblib style Demo
[matploblib](http://matplotlib.org/) [added a style package in 1.4.2](http://matplotlib.org/1.4.2/users/whats_new.html#style-package-added). Here's a quick view of how these styles look.

```python
%matplotlib inline
from matplotlib import style
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(-5, 5, 1000)
ys = np.cos(xs)
print(style.available)
```

## Default

```python
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```

## dark_background

```python
style.use('dark_background')
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```

## bmh

```python
style.use('bmh')
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```

## grayscale

```python
style.use('grayscale')
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```

## ggplot

```python
style.use('ggplot')
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```

```python
## fivethiryeight
```

```python
style.use('fivethirtyeight')
plt.plot(xs, ys, label='$sin(x)$')
plt.legend()
plt.grid()
```
